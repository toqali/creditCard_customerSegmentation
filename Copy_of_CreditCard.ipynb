```python
def percentCustForEachTenure(start, end):
  PerCust_Pay, Nummonth = [], 6
  AllCustomer_ForEach_Tenure = df.groupby("TENURE")['PRC_FULL_PAYMENT'].count()
  CustomerPercentPayment = df[(df['PRC_FULL_PAYMENT'] >= start) & (df['PRC_FULL_PAYMENT'] <= end)].groupby("TENURE")['PRC_FULL_PAYMENT'].count()
  groups = zip(CustomerPercentPayment, AllCustomer_ForEach_Tenure)
  # print the percent of pay in specific Tenure
  for sub, total in groups :
    print(f"Percent of Customer who pay between({start}, {end}) in Tenure ({Nummonth}) = ", round(sub/ total, 2))
    Nummonth += 1
    PerCust_Pay.append(round(sub/ total, 2))
```

```python
print("============= (0.5, 1) ================")
percentCustForEachTenure(0.5, 1)
print("\n============= does not pay ================")
percentCustForEachTenure(0, 0)
print("\n============= Pay the all ================")
percentCustForEachTenure(1, 1)
```

# **Scaling**

```python
from sklearn.preprocessing import StandardScaler
selected_columns = ['BALANCE', 'CREDIT_LIMIT', 'PURCHASES']
selected_features = df[selected_columns]
standard_scaler = StandardScaler()
df_standard_scaled = standard_scaler.fit_transform(selected_features)
```

# **Modeling**

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Calculate inertia for different values of k
ks = range(1, 11)
inertias = []

for k in ks:
    kmeans = KMeans(n_clusters=k, init='k-means++')  # Use KMeans++ initialization
    kmeans.fit(df_standard_scaled)
    inertias.append(kmeans.inertia_)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Inertia vs. Number of clusters')
plt.xticks(ks)
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

# Calculate silhouette scores for different values of k
ks = range(2, 11)
silhouette_scores = []

for k in ks:
    kmeans = KMeans(n_clusters=k)
    cluster_labels = kmeans.fit_predict(df_standard_scaled)
    silhouette_scores.append(silhouette_score(df_standard_scaled, cluster_labels))

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(ks, silhouette_scores, '-o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score vs. Number of clusters')
plt.xticks(ks)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from matplotlib import cm

# Define a range of k values
ks = [2,3, 4]

# Create subplots
fig, axs = plt.subplots(1, len(ks), figsize=(26, 15))

for i, k in enumerate(ks):
    # Initialize KMeans with k clusters
    kmeans = KMeans(n_clusters=k)
    cluster_labels = kmeans.fit_predict(df_standard_scaled)

    # Calculate silhouette scores
    silhouette_avg = silhouette_score(df_standard_scaled, cluster_labels)
    sample_silhouette_values = silhouette_samples(df_standard_scaled, cluster_labels)
    print("Score silhouette :", silhouette_avg)

    y_lower = 10
    for j in range(k):
        # Aggregate silhouette scores for samples in cluster j
        jth_cluster_silhouette_values = sample_silhouette_values[cluster_labels == j]
        jth_cluster_silhouette_values.sort()

        size_cluster_j = jth_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_j

        color = cm.nipy_spectral(float(j) / k)
        axs[i].fill_betweenx(np.arange(y_lower, y_upper),
                              0, jth_cluster_silhouette_values,
                              facecolor=color, edgecolor=color, alpha=0.7)

        axs[i].text(-0.05, y_lower + 0.5 * size_cluster_j, str(j))

        y_lower = y_upper + 10

    axs[i].set_xlim([-0.1, 1])
    axs[i].set_ylim([0, len(df_standard_scaled) + (k + 1) * 10])
    axs[i].set_title('Silhouette plot for k = {}'.format(k))
    axs[i].set_xlabel('Silhouette coefficient values')
    axs[i].set_ylabel('Cluster label')

    axs[i].axvline(x=silhouette_avg, color="red", linestyle="--")

plt.show()
```
